## 3-1 당장 좋은 것만 선택하는 그리디

### 거스름돈

```python
n = 1260
count = 0
coins = [500, 100, 50, 10]

for coin in coins:
    count += n // coin
    n %= coin

print(count)
```

## 3-2 큰 수의 법칙

```python
n, m, k = map(int, input().split())
data = list(map(int, input().split()))

result = 0
data.sort()
fir = data[n-1]
sec = data[n-2]
M = m

while M > 0:
    if M > k:
        result += fir * k
        M -= k
        result += sec
        M -= 1
    else:
        result += fir * M
        break
        
print(result)
```
