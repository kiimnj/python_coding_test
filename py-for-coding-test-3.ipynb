## 3-1 당장 좋은 것만 선택하는 그리디

### 거스름돈

```python
n = 1260
count = 0
coins = [500, 100, 50, 10]

for coin in coins:
    count += n // coin
    n %= coin

print(count)
```

## 3-2 큰 수의 법칙

```python
n, m, k = map(int, input().split())
data = list(map(int, input().split()))

result = 0
data.sort()
fir = data[n-1]
sec = data[n-2]
M = m

while M > 0:
    if M > k:
        result += fir * k
        M -= k
        result += sec
        M -= 1
    else:
        result += fir * M
        break
        
print(result)
```

## 3-3 숫자 카드 게임

```python
n, m = map(int, input().split())
cards = []
for i in range(n):
    cards.append(list(map(int, input().split())))
mins = [min(cards[i]) for i in range(n)]
print(max(mins))
```

- mins에서 처음에 n을 m으로 잘못 적음
- cards = [list(map(int, input().split())) for i in range(n)]로 줄여쓰기도 가능
- 사실상 n으로 같은 반복을 하는 두 반복문이라 합쳐야 했음
- 모범답안은 첫 반복문에서 result = max(result, min_value)와 같이 최대값 연산을 매번 수행
- 모범답안2는 min()에서 원소 두개만 넣고 최소값 연산을 반복문 내에서 또 매번 반복 수행해서 2중 반복

## 3-4 1이 될 때까지

```python
n, k = map(int, input().split())
count = 0
while n != 1: #혹은 n > 1
    if n % k != 0:
        n -= 1
    else:
        n //= k
    count += 1
print(count)
```

- 'if not(oo==oo)' = 'if oo!=oo'
- 내 풀이가 가장 직관적인 초보용 코드
- 모범답안들은 n이 1이 될때까지가 아니고 n<k인 조건으로 루프 돌림
- 그리디는 결과 최적화 중심: 내 코드는 while(한단계씩 체크하며 연산), 모범1은 그리디(큰구간의 연산을 먼저 수행). 빼기 구간 나누기 구간 분리해서 처리, 불필요한 while 제거, 약간의 최적화(if 판단 횟수와 % 연산 횟수). 그치만 둘다 복잡도 O(n). n요소까지는 아니지만 상수만큼의 연산횟수 차이
- 모범2는 1을 빼기하는 연산횟수를 계산하여 한번에 빼기하여 촤적화, 내풀이 복잡도는 O(연산 횟수), 모범2는 O(log n)
