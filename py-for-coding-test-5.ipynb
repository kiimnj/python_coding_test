# 7 이진 탐색

## 7-2 부품 찾기

```python
import sys

n = int(sys.stdin.readline().rstrip())
n_data = list(map(int, sys.stdin.readline().rstrip().split()))
m = int(sys.stdin.readline().rstrip())
m_data = list(map(int, sys.stdin.readline().rstrip().split()))

for i in m_data:
    if i in n_data:
        print("yes", end=' ')
    else:
        print("no", end=' ')
```

- 정렬 및 탐색 알고리즘 미숙지로 출제의도(훈련의도)와 최적화 하나도 안한 코드 작성, n, m 값은 쓰지도 않음 (근데 계수정렬만 n쓰고 다른풀이들도 안쓰긴함)
- `m_data`와 `n_data` 위치 주의
- 문제 직전에 `sys`가르치길래 쓴건데 코드만 길어짐. 100만정도는 input으로도 충분하고 list대신 set을 쓰는게 오히려 훨씬 시간 줄어든다고 함
- 이진탐색 사용시 탐색함수를 구현하여 반복문 내 if문의 in에 함수 사용
- 계수정렬 사용시 계수 인덱스리스트를 구현하여 입력받으며 존재여부를 저장하고 반복문 내 if문에 리스트값 사용
- `set()` 사용 시 입력받을 때에 이용(함수가 아니고 자료형!): `n_data=set(map(int, input().split()))`, 그냥 리스트 사용 시 시간초과 가능성
