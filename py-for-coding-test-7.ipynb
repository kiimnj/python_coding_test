- 아래 모범답안의 ##달아놓은 부분이 이진탐색 코드와 모범부분이 다른코드(조건이 적용된) 부분, 그 아래의 이진탐색 코드와 비교

```python
# 모범답안
n, m = list(map(int, input().split(' '))) #공백 한칸만 기준
array = list(map(int, input().split())) #모든 공백, 줄바꿈이 기준

# 이진 탐색 수행 (반복적)
start = 0
end = max(array)
result = 0
while(start <= end):
    mid = (start + end) // 2

    total = 0 ## 내가 생각한 조건이 들어간 부분, h 대신 탐색값 mid가 변수
    for x in array: ##
        if x > mid: ##
            total += x - mid ##

    # 떡볶이 양이 부족한 경우 더 많이 자르기 (H 탐색 범위값 줄이기)
    if total < m: ##
        end = mid - 1
    # 떡볶이 양이 충분한 경우 덜 자르기 (H 탐색 범위값 키우기)
    else:
        result = mid ## 진행될수록 최적화될뿐 가능한 값이므로 result에 기록
        start = mid + 1

print(result)
```

```python
# 이진 탐색 소스코드 구현 (반복문)
def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        # 찾은경우 중간점인덱스 반환 후 탐색종료, 모범코드는 최적화이므로 결과기록 이후에도 탐색진행
        if array[mid] == target: ##
            return mid ##
        
        # 중간점의 값보다 찾고자 하는 값이 작은 경우 왼쪽 확인
        elif array[mid] > target:
            end = mid - 1
        # 중간점의 값보다 찾고자 하는 값이 작은 경우 오른쪽 확인
        else:
            start = mid + 1
    return None
```
